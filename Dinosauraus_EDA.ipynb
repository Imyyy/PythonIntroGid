### Exploring summary statistics
Idea here is to start looking around the data, working out what information is available and what we can pull out.

```python
print(data.columns)
print(data.shape)
```

```python
data.head()
```

The columns x, y, dino aren't too interesting, perhaps we'd want to rename them and come up with something more creative? E.g. dino height, dino length?

```python
print(data['dataset'].unique())
```

```python
# Looking at the mean of each column - very similar
data.groupby('dataset')[['x', 'y']].mean()
```

```python
# Looking at the data
data.groupby('dataset').agg(
    x_mean=('x', np.mean),
    y_mean=('y', np.mean),
    x_median =('x', np.median),
    y_median =('y', np.median))
```

Means all similar, little variations in the median. So lets plot to work out where to go next

### Plots
AIM: Work out where the interesting plots are

```python
# Plotting all data to work out if anything interesting is happening
plt.scatter(data['x'], data['y'])
plt.show()
```

```python
# For each categorical, matplotlib scatter plot it so we can see what it gives us
shapes = data['dataset'].unique()
print(shapes)

for shape in shapes:

  # Filter data for just one geometry of plot
  subset = data[data['dataset'] == shape]
  subset.reset_index(inplace=True)

  # Plot data subset
  plt.scatter(subset['x'], subset['y'])
  plt.title(shape)
  plt.show()
```

```python
# Splitting out just the dinosaur plot
dino = data[data['dataset'] == "dino"]

print(dino['x'].mean())
print(dino.shape)

plt.scatter(dino['x'], dino['y'])
plt.show()
```

```python
# Can we rainbowify the dinosaur?
plt.scatter(dino['x'], dino['y'], c = dino["y"], cmap="rainbow")
plt.show()
```
